```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
import statsmodels.formula.api as smf
import pingouin as pg
```

## a)

```python
# Define the data
data = pd.DataFrame({'revenue': pd.Series([3, 8, 19, 22, 31, 42, 48, 52, 54, 61]),
                     'area': pd.Series([150, 180, 420, 480, 660, 1000, 1300, 1500, 1600, 1710])})
```

```python
sns.scatterplot(y = 'revenue', x= 'area', data=data)
```

## b)

```python
data.revenue.corr(data.area)
```

## c)

```python
data['revenue_dollar'] = data.revenue * 1.4
data.revenue_dollar.corr(data.area)
```

## d)

```python
model = smf.ols('revenue ~ area', data=data).fit()
model.summary()
```

## e)

```python
pg.qqplot(model.resid)
```
