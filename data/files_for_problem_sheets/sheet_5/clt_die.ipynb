# Illustrate the Central Limit Theorem using a die

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats

sns.set_theme()
```

```python
# Simulate 10000 dice rolls and store them in a matrix (1 row per sequence).
W = stats.randint.rvs(low=1, high=7, size=(1000,10), random_state=452)
print(W, W.shape)
```

```python
# Compute the row-wise mean
MW = np.mean(W, axis=1)
```

```python
# Histogram
sns.displot(MW)
None
```

Since the die can only show a limited number of values, the distribution of the means for $n=10$ is still quite far from a normal distribution.

```python
# Same simulation but with a length n = 50
W = stats.randint.rvs(low=1, high=7, size=(1000,50), random_state=452)
print(W, W.shape)

# Compute the row-wise mean
MW50 = np.mean(W, axis=1)

# Histogram
sns.displot(MW50)
None
```

```python
# Same simulation but with a length n = 100
W = stats.randint.rvs(low=1, high=7, size=(1000,100), random_state=452)
print(W, W.shape)

# Compute the row-wise mean
MW100 = np.mean(W, axis=1)

# Histogram
sns.displot(MW100)
None
```

```python
# Plotting every graph together
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.histplot(MW, ax=axes[0], kde=False)
axes[0].set_title('n = 10')

sns.histplot(MW50, ax=axes[1], kde=False)
axes[1].set_title('n = 50')

sns.histplot(MW100, ax=axes[2], kde=False)
axes[2].set_title('n = 100')

plt.tight_layout()
plt.show()
```
